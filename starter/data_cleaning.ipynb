```python
import pandas as pd
```

```python
df = pd.read_csv('data/census.csv')
```

```python
df.sample(50)
```

```python
for col in df.select_dtypes('object').columns:
    df[col] = df[col].str.strip()

df.columns = df.columns.str.strip()
```

```python
df.to_csv('data/census_clean.csv', index=False)
```
